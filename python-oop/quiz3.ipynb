```python
#8) Aşağıda yazdırılan sınıfı incelediğinizde kedim.yasiCarp() kodunun çıktısı nedir
```

```python
class Kedi():

    def __init__(self,isim,yas=5):
        self.isim = isim
        self.yas = yas

    def yasiCarp(self):
        return self.yas * 3
```

```python
kedim = Kedi("Tonton")
```

```python
kedim.yasiCarp()
```

```python
#9) Aşağıdaki kodun çıktısı ne olacaktır?
```

```python
class Ogrenci(): #encapsulation

    def __init__(self,isim,sinavNotu):
        self.isim = isim
        self.__sinavNotu = sinavNotu

    def notuGoster(self):
        print(f"{self.isim} sınav notu: {self.__sinavNotu}")
```

```python
ogrenci = Ogrenci("Mehmet",85)
```

```python
ogrenci.__sinavNotu = 75
```

```python
ogrenci.notuGoster()
```
