```python
class Musician():

    def __init__(self,name):
        self.name = name
        print("musician class")

    def test1(self):
        print("test1")
    def test2(self):
        print("test2")
```

```python
atil = Musician("Atil")
```

```python
atil.name
```

```python
atil.test1()
```

```python
atil.test2()
```

```python
class MusicianPlus(Musician):
    def __init__(self,name):
        Musician.__init__(self,name)
        print("musician plus")

    def test3(self):
        print("test3")
    #override
    def test1(self):
    
        print("test1 test1 test1")
        
```

```python
atlas = MusicianPlus("Atlas")
```

```python
atlas.test1()
```

```python
atlas.test2()
```

```python
atlas.name = "Atlas Samancıoğlu"
```

```python
type(atlas)
```

```python
atlas.test3()
```

```python
class Banana():
    def __init__(self,name):
       self.name = name

    def info(self):
        return f"100 calories {self.name}"

class Apple():

    def __init__(self,name):
         self.name = name
    def info(self):
        return f"150 calories {self.name}"
        
    

     
```

```python
banana = Banana("banana")
```

```python
apple = Apple("apple")
```

```python
banana.info()
```

```python
apple.info()
```

```python
fruitList = [banana,apple]
```

```python
for fruit in fruitList:
   print(fruit.info())
```

#ENCAPSULATİON

```python
class Phone():

    def __init__(self,name,price):
        self.name = name
        self.__price = price
    def info(self):
        print(f"{self.name} price is: {self.__price}")
    def changePrice(self,price):
        self._price = price
```

```python
iphone = Phone("Iphone 14",500)
```

```python
iphone.info()
```

```python
iphone.__price = 400
```

```python
iphone.info()
```

```python
iphone.changePrice(300)
```

```python
iphone.info()
```

ABSTRACTİON

```python
from abc import ABC, abstractmethod
```

```python
class Car(ABC):

    @abstractmethod
    def maxSpeed(self):
        pass
        
```

```python
#myCar = Car()
```

```python
class Tesla(Car):

    def maxSpeed(self):
        print("200km")
```

```python
tesla = Tesla()
```

```python
tesla.maxSpeed()
```

```python
class Mercedes(Car):
    def maxSpeed(self):
        print("250km")
```

```python
mercedes = Mercedes()
```

```python
mercedes.maxSpeed()
```

```python
class Fruit():

    def __init__(self,name,calories):
        self.name = name
        self.calories = calories

    def __str__(self):
        return f"{self.name}: {self.calories} calories"

    def __len__(self):
        return self.calories
        
```

```python
myFruit = Fruit("banana",150)
```

```python
myFruit.calories
```

```python
myFruit.name
```

```python
print(myFruit)
```

```python
len(myFruit)
```

```python
class Train():

    def __init__(self,name):
        self.name = name

    def __getitem__(self,key):
        if key == "a":
           return self.name
        else:
            return "Not found"
```

```python
myTrain = Train("myTrain")
```
