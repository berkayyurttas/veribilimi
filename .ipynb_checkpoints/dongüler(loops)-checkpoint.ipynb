```python
my_list = [10,20,30,40,50,60,70]
```

```python
print("for loop started")
for num in my_list:
    print(num)
print("for loop ended")
```

```python
40 in my_list
```

```python
for number in my_list:
    print(number)
    if number == 40:
        
        print("yes")
        break
        
```

```python
my_list = [10,20,30,40,50,60,70]
```

```python
for number in my_list:
   
    if number == 40:
        continue
    print(number)
       
       
```

```python
x = 0
```

```python
print("while loop started")
while x < 10:
    print(x)
    x = x + 1
print("while loop ended")
```

```python
last_list = [10,20,30,10,50]
```

```python
while 20 in last_list:
    print("20 in last_list")
    last_list.pop()
```
