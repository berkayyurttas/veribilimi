```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
 data1 = np.linspace(0,10,20)
```

```python
data1
```

```python
data2 = data1 ** 2 
```

```python
plt.subplots()
```

```python
my_fig, my_axes = plt.subplots()
```

```python
my_fig, my_axes = plt.subplots()
my_axes.plot(data1, data2,"green")
my_axes.plot(data2, data1,"black")
plt.show()
```

```python
(new_fig, new_axes) = plt.subplots()
new_axes.plot(data1, data1 + 2, color = "blue", linewidth=1.0)
new_axes.plot(data1, data1 + 3, color = "red", linewidth=4.0)
new_axes.plot(data1, data1 + 4, color = "brown", linestyle = ":")
new_axes.plot(data1, data1 + 5, color = "yellow", linestyle= "--")

new_axes.plot(data1, data1 + 6, color = "#000000", linestyle= "--", marker="o", markersize= 8, markerfacecolor="red")

new_axes.plot(data1, data1 + 7, color = "#000000", linestyle= "--", marker="+", markersize= 9, markerfacecolor="grey")


plt.show()
```
