```python
age_list = [10,20,30,30,30,40,50,60,70,75]
```

```python
weight_list = [20,60,80,85,87,70,90,95,90,100]
```

```python
len(age_list)
```

```python
plt.plot(age_list, weight_list, "b") #yellow, g, b, r, b blue
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.savefig("test.png")
plt.show()
```

```python
plt.plot(np_age_list, np_weight_list, "g")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
numpy_arr1 = np.linspace(0,10,20)
```

```python
numpy_arr1
```

```python
numpy_arr2 = numpy_arr1 ** 3
```

```python
numpy_arr2
```

```python
plt.plot(numpy_arr1, numpy_arr2)
plt.show()
```

```python
plt.plot(numpy_arr1, numpy_arr2, "r*-") #yldzla data pointleri grrz
plt.show()
```

```python
plt.plot(numpy_arr2, numpy_arr1,"b*-")
```

```python
plt.subplot(1,2,1)
plt.plot(numpy_arr1, numpy_arr2, "r*-")
plt.subplot(1,2,2) #1row 2 column 2.grafik
plt.plot(numpy_arr2, numpy_arr1, "g--")
plt.show()
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.6,0.6])
figure_axes.plot(numpy_arr1, numpy_arr2, "g")
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Axis")
figure_axes.set_title("Graph Title")
plt.show()
```

```python
new_fig = plt.figure(dpi=100)

new_axes = new_fig.add_axes([0.1,0.1,0.9,0.9])
new_axes.plot(numpy_arr1, numpy_arr1 ** 2, label = "numpy array **2")
new_axes.plot(numpy_arr1, numpy_arr1 ** 3, label = "numpy array **3")
new_axes.legend(loc=0) #loc0 labeln nerde durcana bakar, ** sl demek

plt.show()
```
